```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(42)

# Define note types
note_types = ['progress', 'nursing', 'discharge', 'radiology']

# Function to generate slightly different Gaussian distributions
def generate_distributions(mean, std, size=1000):
    return np.random.normal(loc=mean, scale=std, size=size)

# Generate 3 sets of 4 distributions
distributions_sets = []
for _ in range(3):
    distributions = {
        'progress': generate_distributions(mean=0, std=1),
        'nursing': generate_distributions(mean=1, std=1.2),
        'discharge': generate_distributions(mean=0.5, std=0.8),
        'radiology': generate_distributions(mean=-1, std=1.5)
    }
    distributions_sets.append(distributions)

# Plot the distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Define colors for the distributions
colors = {
    'progress': 'blue',
    'nursing': 'green',
    'discharge': 'red',
    'radiology': 'purple'
}

for i, distributions in enumerate(distributions_sets):
    for note_type, data in distributions.items():
        sns.histplot(data, kde=True, ax=axes[i], color=colors[note_type], 
                     label=note_type, stat="density", common_norm=False, 
                     binwidth=0.2, alpha=0.6)
    axes[i].set_title(f'Distribution Set {i+1}')
    axes[i].legend(title='Note Type')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Density')

plt.tight_layout()
plt.show()
```
